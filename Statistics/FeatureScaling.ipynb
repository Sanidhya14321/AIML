```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
df.head()
```

```python
df.describe().round(2)
```

```python
new_df = pd.DataFrame(df, columns=['Age' , 'Tenure'])
```

```python
new_df.head()
```

```python
new_df['Age'] = new_df['Age'].fillna(new_df['Age'].mean())
```

```python
scaler = MinMaxScaler()
normalized_df = scaler.fit_transform(new_df)
print(normalized_df)
```

```python
x_array = np.array([[2],[3],[5],[6],[6]])
scaler = MinMaxScaler()
normalized_array = scaler.fit_transform(x_array)
mn = x_array.min()
mx = x_array.max()
print(mn, mx, '\n', normalized_array)
```

```python
x_array = np.array([[2],[3],[5],[6],[6]])
scaler = StandardScaler()
normalized_array = scaler.fit_transform(x_array)
s = x_array.std()
m = x_array.mean()
print(s , m,'\n',  normalized_array)
```
